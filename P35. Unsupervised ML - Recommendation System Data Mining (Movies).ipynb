#  Unsupervised ML - Recommendation System Data Mining (Movies)

```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# Import Dataset
movies=pd.read_csv('Database/Movie.csv')
movies
```

```python
# Sort by User IDs
movies.sort_values(['userId'])
```

```python
# number of unique users in the dataset
len(movies['userId'].unique())
```

```python
# number of unique movies in the dataset
len(movies['movie'].unique())
```

```python
# converting long data into wide data using pivot
movies2=movies.pivot_table(index='userId',columns='movie',values='rating').reset_index(drop=True)
movies2
```

```python
# Replacing the index values by unique user Ids
movies2.index=movies['userId'].unique()
movies2
```

```python
# Impute those NaNs with 0 values
movies2.fillna(0,inplace=True)
movies2
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(movies2.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids 
user_sim2.index=movies['userId'].unique()
user_sim2.columns=movies['userId'].unique()
user_sim2
```

```python
# Slicing first 5 rows and first 5 columns
user_sim2.iloc[0:5,0:5]
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim, 0)
user_sim2.iloc[0:5,0:5]
```

```python
# Most Similar Users
user_sim2.idxmax(axis=1)[0:5]  # (idxmax) Returns index of the maximum element
```
